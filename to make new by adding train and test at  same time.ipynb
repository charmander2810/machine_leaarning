```python
house_train=house_train.drop(["SalePrice"],axis=1)
house_train.head()
```

```python
#fill the empty space of integer using mean value    train
for label,content in house_train.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            house_train[label]=content.fillna(content.median())
```

```python
#fill the empty space of integer using mean value   test
for label,content in house_test.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            house_test[label]=content.fillna(content.median())
```

```python
#fill the empty space of string using missing train
for label,content in house_train.items():
    if pd.api.types.is_string_dtype(content):
        if pd.isnull(content).sum():
            house_train[label]=content.fillna("missing")
```

```python
#use labelencode
import pandas as pd
from sklearn.preprocessing import LabelEncoder

def Encoder(df):
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

```python
df_trains = Encoder(house_train)
```

```python
df_tests=Encoder(house_test)
```

```python
df_trains
```

```python
df_tests
```

```python
# its time for creating a model
#model
np.random.seed(40)
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df_trains,
                                               y,
                                               test_size=0.1)
model=LinearRegression().fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
final_solution=model.predict(df_tests)
```
